```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.pyplot import figure
from matplotlib import rc
from matplotlib import font_manager
mpl.rcParams['axes.linewidth'] = 2 #set the value globally
font_dirs = ['/scratch/user/peterchao1/font/']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    font_manager.fontManager.addfont(font_file)
mpl.rcParams["font.family"] = "Calibri"
```

```python
OHE_LC_train_size = np.load('OHE_LC_train_size.npy')
OHE_LC_train_loss = np.load('OHE_LC_train_loss.npy')
OHE_LC_val_loss = np.load('OHE_LC_val_loss.npy')
OHE_LC_train_score = np.load('OHE_LC_train_score.npy')
OHE_LC_val_score = np.load('OHE_LC_val_score.npy')
```

```python
plt.plot(OHE_LC_train_size.mean(1), OHE_LC_train_loss.mean(1), "o-", color="#4575b4", label="train_loss")
plt.plot(OHE_LC_train_size.mean(1), OHE_LC_val_loss.mean(1), "o-", color="#d73027", label="val_loss")
plt.ylabel('RMSE ('+r'$\AA$)')
plt.xlabel('Train size')
plt.legend()
plt.title('One-Hot Encoding Linear Regression Learning Curve')
plt.savefig('OHE_LRR_LC.png',dpi=400)
```

```python
plt.plot(OHE_LC_train_size.mean(1), OHE_LC_train_score.mean(1), "o-", color="#4575b4", label="train_loss")
plt.plot(OHE_LC_train_size.mean(1), OHE_LC_val_score.mean(1), "o-", color="#d73027", label="val_loss")
plt.ylabel(r'$R^2$')
plt.xlabel('Train size')
plt.legend()
plt.title('One-Hot Encoding Linear Regression Learning Curve')
plt.savefig('OHE_LRR_LC_score.png',dpi=400)
```
