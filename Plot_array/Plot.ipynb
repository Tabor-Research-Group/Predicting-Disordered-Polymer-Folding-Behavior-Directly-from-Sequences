```python
def coeff_determination(y_true, y_pred):
    SS_res =  sum((y_true-y_pred)**2)
    SS_tot = sum((y_true-sum(y_true)/len(y_true))**2)
    return ( 1 - SS_res/(SS_tot) )

def percent_error(y_true, y_pred):
    return sum(abs(y_pred-y_true)/y_true)/len(y_true)

def MAE(y_true,y_pred):
    return sum(abs(y_true-y_pred))/len(y_pred)

def MSE(y_true,y_pred):
    return sum((y_true-y_pred)**2)/len(y_pred)

def RMSE(y_true,y_pred):
    return sqrt(sum((y_true-y_pred)**2)/len(y_pred))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.decomposition import PCA
from collections import Counter
import random
from sklearn.preprocessing import MinMaxScaler
```

```python
seq = np.load("../New_data_nu.npy",allow_pickle=True)
seq = seq.item()
data_seq = []
Rg = []
nu = []
for i in range(len(seq)):
    data_seq.append(seq[i][0])
    Rg.append(seq[i][1][1][1])
    nu.append(seq[i][1][1][3])
Rg = np.array(Rg).reshape(-1,1)
nu = np.array(nu).reshape(-1,1)
Y = np.hstack((Rg,nu))
tokenizer = Tokenizer(char_level=True)
tokenizer.fit_on_texts(data_seq)
sequences = tokenizer.texts_to_sequences(data_seq)
one_hot_results = tokenizer.texts_to_matrix(data_seq, mode='count')
one_hot_results = np.delete(one_hot_results,[0],axis=1)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
X = one_hot_results
```

```python
def LC_split_CL(fold,split,seed,c):
    fold = fold
    split = split
    total_split = fold*split
    Train_indices = [[] for i in range(total_split)]
    Test_indices = []
    for i in range(len(list(c.keys()))):
        random.seed(seed)
        l = list(c.values())[i]
        Total_indices = np.arange(0,l,1).tolist()
        random.shuffle(Total_indices)
        Total_train_size = 0.8*l
        for j in range(total_split):
            tmp_train = np.array(Total_indices[int(Total_train_size/total_split*j):int(Total_train_size/total_split*(j+1))])
            for k in tmp_train:
                Train_indices[j].append(k+i*l)
        tmp_test = np.array(Total_indices[int(Total_train_size):int(l)])
        for k in tmp_test:
            Test_indices.append(k+i*l)
    return Train_indices, Test_indices

def get_CL_OHA(seqs):
    CL = []
    for i in range(len(seqs)):
        CL.append(sum(seqs[i]))
        c = Counter(CL)
    return c

fold = 6
seed = 10
split = 8
c = get_CL_OHA(X)
Train_indices,Test_indices = LC_split_CL(fold,split,seed,c)
```

```python
X_train_unscaled = []
X_test_unscaled = []

for i in Train_indices:
    X_train_unscaled.append(X[i])

for i in Test_indices:
    X_test_unscaled.append(X[i])

    
X_train_unscaled = np.vstack(X_train_unscaled)
X_test_unscaled = np.vstack(X_test_unscaled)

scaler = MinMaxScaler()
scaler.fit(X_train_unscaled)
X_train = scaler.transform(X_train_unscaled)
X_test = scaler.transform(X_test_unscaled)
```

```python
pca_OHA = PCA(n_components=20)
pca_OHA.fit(X_train_unscaled)
X_train_pca = pca_OHA.transform(X_train_unscaled)
X_test_pca = pca_OHA.transform(X_test_unscaled)
```

```python
pca_accum = []
for i in range(len(pca_OHA.explained_variance_ratio_)+1):
    pca_accum.append(sum(pca_OHA.explained_variance_ratio_[0:i]))
```
